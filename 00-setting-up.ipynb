# Data Visualization

```python
# prepare some data
y = [6, 7, 2, 4, 5]
```

```python
import matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(y)
plt.show()
```

```python
# create a figure 
plt.figure(figsize=(10, 5)) 

# add a line renderer with legend and line thickness
plt.plot(y)

# also need to make the figure readable -- very important from stats perspective!
plt.title("First Example Plot")
plt.xlabel("x vals (s)")
plt.ylabel("y vals (m)")
plt.grid()

plt.show()
```

```python
da['BMXWT'].size
```

```python
## Plotting Weights 
# create a figure 
plt.figure(figsize=(20, 5)) 

# add a line renderer with legend and line thickness
x =  range(da['BMXWT'].size)
y =  da['BMXWT']

plt.scatter(x, y, marker='x', s=0.5)

# also need to make the figure readable -- very important from stats perspective!
plt.title("Weights of Individuals")
plt.xlabel("ID of Individual")
plt.ylabel("Weight (kg)")

plt.show()
```
